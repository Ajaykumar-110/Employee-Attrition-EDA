```python
plt.figure()
sns.countplot(x="Attrition", data=df)
plt.title("Employee Attrition Count")
plt.show()
plt.figure(figsize=(10, 6))
```

```python
plt.figure()
sns.histplot(df["Age"], bins=10, kde=True)
plt.title("Age Distribution of Employees")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df["MonthlyIncome"])
plt.title("Monthly Income Outliers")
plt.show()
```

```python
plt.figure()
sns.countplot(x="Department", hue="Attrition", data=df)
plt.title("Attrition by Department")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure()
sns.countplot(x="JobSatisfaction", hue="Attrition", data=df)
plt.title("Attrition vs Job Satisfaction")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="Attrition", y="YearsAtCompany", data=df)
plt.title("Years at Company vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Key Insights:")
print("- Employees with low job satisfaction show higher attrition")
print("- Attrition is higher in early years at the company")
print("- Sales department has higher attrition")
print("- Higher income employees tend to stay longer")
```
